## 0. Matplotlib
- 파이썬 기반 데이터 분석 환경에서 가장 일반적으로 사용되는 시각화 라이브러리

## 1. `plt.plot(y)`

- 기본 : 라인 플롯(line plot)
    
- data가 시간, 순서등에 따라 변화하는 흐름 파악

- x가 주어지지 않을 경우 자동 생성: `x=range(len(y))`

```python
# data 생성: 5~6개의 item을 가지는 list
# plot(data): 기본적으로 y축의 값으로 data를 mapping
# x > len(data) 만큼의 index number 생성
# 기본 plot = line plot

plt.plot([1, 3, 4, 2, 6, 10])
plt.show()
```

```python
# x tick(x축 value) 설정
# plt.plot(x, y) > list x와 list y의 length가 같아야 함
plt.plot([2, 4, 6, 8, 10, 12], [1, 3, 4, 2, 7, 10])
plt.show()
```

- 축 이름과 plot 제목 지정

1. plot 제목: `plt.title('title')`
2. x축 이름: `plt.xlabel('x_name')`
3. y축 이름: `plt.ylabel('y_name')`

```python
list_x = [-12, -10, -8, -6, -4, -2]
list_y = [1, 3, -4, -2, 6, -10]

plt.plot(list_x, list_y)

plt.title("라인 그래프 1")
plt.xlabel('axis x')
plt.ylabel('axis y')

plt.show()
```

## 2. style assignment

- `plt.plot([x], y, [fmt], **kwargs)`

    - `fmt=[color][marker][line]`

        - 색상, 마커, 선 등에 대한 포맷 옵션 설정

<br>

1. 마커 참조 도큐먼트   
https://matplotlib.org/3.1.3/api/markers_api.html

<br>

2. 색상 참조 도큐먼트   
https://matplotlib.org/3.1.0/gallery/color/named_colors.html

<br>

3. 라인 스타일 참조 도큐먼트   
https://matplotlib.org/gallery/lines_bars_and_markers/line_styles_reference.html

```python
# style string insert: [fmt] = color + marker + line
# item 생략 시 > defalut value적용
# 'o': line 없이 큰 점(원 모양)으로 data 위치 표시
plt.plot([1, 2, 3, 4], [2, 6, 10, 7], 'o')
plt.show()
```

```python
# o-: (color=default), marker=circle, line=실선
# fmt[color][marker][line]
plt.plot(list_x, list_y, 'o-')
plt.show()
```

```python
# o--: marker=circle, line=dash line 
plt.plot(list_x, list_y, 'o--')
plt.show()
```

```python
# `ro:` > color=red, marker=circle, line=colon line
plt.plot(list_x, list_y, 'ro:')
plt.show()
```

<br>

- 기타 스타일 파라미터 종류
    - `color='magenta'`: 색상
    - `linestyle='-.'`
    - `marker=^`
    - `markerfacecolor='orange'`
    - `markersize='20'`

```python
plt.plot([1, 2, 3, 4], [2, 6, 10, 7],
        color = 'magenta', linestyle='-.', marker='^',
        markeredgecolor='r', markerfacecolor='orange', markersize='20')
plt.show()
```

## 3. 축 범위 지정

### 3.1. `matplotlib.pyplot.axis()`

- xmin, xmax, ymin, ymax = `plt.axis([xmin, xmax, ymin, ymax])`
    - xmin : x축 최소값
    - xmax : x축 최대값
    - ymin : y축 최소값
    - ymax : y축 최대값
    
### 3.2. `matplotlib.pyplot.xlim`, `matplotlib.pyplot.ylim`

- `plt.xlim(min, max)`
- `plt.ylim(min, max)`

```python
# default: asix=x의 min, max, asix=y의 min, max에 맞춰진 graph
plt.plot([1, 2, 3, 4], [10, 3, 8, 13])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [10, 3, 8, 13])
plt.axis([0, 5, 0, 16])
plt.show()
```

```python
# lind에 비해 지나치게 작은 plot을 구성하면 part of plot이 잘려나간다.

plt.plot([1, 2, 3, 4], [10, 3, 7, 13])
plt.axis([2, 4, 4, 10])
plt.show()
```
