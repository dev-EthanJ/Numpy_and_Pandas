
# Data visualization with matplotlib 데이터 시각화

## 0. Matplotlib

- 파이썬 기반 데이터 분석 환경에서 가장 일반적으로 사용되는 시각화 라이브러리

- 하위 라이브러리인 `pyplot`을 통해 주요 기능을 사용

- `plotly`, `seaborn` 등과 같은 라이브러리가 최근 경쟁력을 갖추고 있음

- pandas 객체의 기본적인 시각화 도구

- `plot()` method를 통해 데이터를 시각화

- 종류

    - 라인 플롯(line plot)

    - 바 차트(bar chart)

## 1. `matplotlib.pyplot.plot(y)`: line plot

- `plt.plot(y)`: 기본적으로 y축의 값으로 data `y`를 mapping
        
    - x축: `len(y)`를 기반으로 한 index number 생성: `range(len(y))`
    
    - default: line plot

```python
data_list = np.random.randint(1, 10, (6))
plt.plot(data_list)
plt.show()
```

<br>

- `x tick` x축 value 설정

    - `plt.plot([x tick list], data)`: x tick list 길이와 data의 length가 같아야 함

```python
plt.plot([2, 4, 6, 8, 10, 12], np.random.randint(1, 10, (6)))
plt.show()
```

- 축 이름과 plot 제목 지정

<br>

- 두 개의 선을 그리기

    - 각각의 라인 그래프를 순차적으로 `plt.plot()` command로 입력

```python
x_tick = list()

for x in range(-10, 10, 2):
    x_tick.append(x)

plt.plot(x_tick, np.random.randint(-10, 11, (len(x_tick))))
plt.plot(x_tick, np.random.randint(-10, 11, (len(x_tick))))
    
plt.xlabel('value')
plt.ylabel('time')
plt.title('line plot example 2')

plt.show()
```

<br>

- 범례(legend) 추가: `plt.legend([list of legend str])`

```python
plt.plot(x_tick, np.random.randint(-10, 11, (len(x_tick))))
plt.plot(x_tick, np.random.randint(-10, 11, (len(x_tick))))
    
plt.xlabel('axis=1')
plt.ylabel('axis=0')
plt.title('line plot example 3')
plt.legend(['data A', 'data B'])
plt.show()
```

## 2. style assignment

- `plt.plot([x], y, 'fmt', **kwargs)`

    - `'fmt'=[color][marker][line]`

        - 색상, 마커, 선 등에 대한 포맷 옵션 설정

1. 마커 참조 도큐먼트   
https://matplotlib.org/3.1.3/api/markers_api.html

2. 색상 참조 도큐먼트   
https://matplotlib.org/3.1.0/gallery/color/named_colors.html

3. 라인 스타일 참조 도큐먼트   
https://matplotlib.org/gallery/lines_bars_and_markers/line_styles_reference.html

<br>

- style string : `'fmt' = 'color' + 'marker' + 'line'`
    
    - string item 생략 시: default style 적용
        
<br>        
        
- fmt=`o`: line없이 큰 점(원 모양)으로 data 표시

```python
x_tick = list()

for x in range(1, 6):
    x_tick.append(x)

y_data = np.random.randint(1, 6, len(x_tick))

plt.plot(x_tick, y_data, 'o')

plt.show()
```

<br>

- str `'fmt'` = color + marker + line

    - `'o-'`: color=default, marker:`o`=`circle`, line:`-`=`실선`

```python
plt.plot(x_tick, y_data, 'o-')

plt.show()
```

<br>

- `'o--'`: color=default, marker:`o`=circle, line:`--`=dash_line

```python
plt.plot(x_tick, y_data, 'o--')

plt.show()
```

<br>

- `'ro:'`: color:`r`=red, marker:`o`=circle, lind:`:`=점으로 된 점선

```python
plt.plot(x_tick, y_data, 'ro:')

plt.show()
```

<br>

- 기타 스타일 파라미터 종류
    - `color='magenta'` > 선 색상
    - `linestyle='-.'` > - + 점선
    - `marker=^` > 삼각형 marker
    - `markerfacecolor='orange'` > marker 색상
    - `markersize='20'` > marker 크기

```python
plt.plot(x_tick, y_data, color = 'magenta', linestyle='-.', marker='^',
        markeredgecolor='r', markerfacecolor='orange', markersize='20')

plt.show()
```

## 3. 축 범위 지정

### 3.1. `matplotlib.pyplot.axis([x_m, x_M, y_m, y_M])`

- `plt.axis([xmin, xmax, ymin, ymax])`
    - xmin : x축 최소값
    - xmax : x축 최대값
    - ymin : y축 최소값
    - ymax : y축 최대값
    
### 3.2. `matplotlib.pyplot.xlim(m, M) or ylim(m, M)`

- `plt.xlim(min, max)`

- `plt.ylim(min, max)`

<br>

- default: `axis=x`의 min, max, `axis=y`의 min, max에 맞춰진 graph

```python
axis_x = [1, 2, 3, 4] 
axis_y = [10, 4, 12, 8]

plt.plot(axis_x, axis_y)

plt.show()
```

<br>

- `plt.axis([x_min, x_Max, y_min, y_Max])`

```python
plt.plot(axis_x, axis_y)
plt.axis([-4, 8, 0, 16])

plt.show()
```

<br>

- `plot` line에 비해 지나치게 작은 `plt.axis()`를 설정하면 일부 `plot`이 잘린다.

```python
plt.plot([1, 2, 3, 4], [17, 2, 12, 14])
plt.axis([1, 4, 4, 12])

plt.show()
```

## 4. `matplotlib.pyplot.scatter(x, y)`: scatter plot

- 대표값을 중심으로 data가 흩어져 있는 정도를 시각화

- data의 분산, data간의 상관관계 파악 가능

- `plt.scatter(x, y, s=None, c=None, data=None)`

    - `s`: marker size
    
    - `c`: marker color
    
    - `data`: `dict` 자료형 사용하여 데이터 전달 시 keyword string을 통해 data 사용 

<br>

- 정규분포 확률밀도함수에서 100개의 표본 추출

    - `np.random.normal()`

```python
data_x = np.random.normal(0, 1, (100))
data_y = np.random.normal(0, 1, (100))

plt.scatter(data_x, data_y)

plt.show()
```

```python
print(data_x[0], data_y[0])
```

```python
plt.plot(data_x, data_y)

plt.show()
```

<br>

- optional parameter: `c=''`: 점 색상, `s= `: 점 크기

```python
plt.scatter(data_x, data_y, c='gold', s=100)

plt.show()
```

## 5. `matplotlib.pyplot.bar(x, y)`: Bar chart

- 여러 값의 상대적인 크기 비교에 적합, 원형 차트보다 일반적으로 사용됨

- `plt.bar(x, y)`

    - `x` data = category value > `bar()`, `barh()`로 Bar chart 시각화
    
        - x축 Bar chart: `plt.bar(x_tick, y_data)`
        
        - y축 Bar chart: `plt.barh(y_tick, x_data)`

```python
data_x = ['김', '이', '박']
data_y = [2, 10, 5]

plt.bar(data_x, data_y)

plt.show()
```

```python
plt.barh(data_x, data_y)

plt.show()
```

<br>

- `plt.title('plot title')`

- `plt.xlabel('axis=x label')`

- `plt.ylabel('axis=y label')`

- `plt.xticks(x, [new_tick_list])`

```python
plt.bar(data_x, data_y)
plt.title('Bar Chart')

plt.xlabel('성씨')
plt.ylabel('인구 수')

plt.xticks(data_x, ['Kim', 'Lee', 'Park'])
plt.show()
```

<br>

- `plt.bar(x, y)` object: 객체화(변수 저장) 가능

    - object indexing: 각각의 `bar`마다 `obj.set_color('color')`: 색상 변경 가능

```python
bar_list = plt.bar(data_x, data_y)

color_list = ['r', 'g', 'b']
i = 0

for bar in bar_list:
    bar.set_color(color_list[i])
    i += 1

plt.show()
```
