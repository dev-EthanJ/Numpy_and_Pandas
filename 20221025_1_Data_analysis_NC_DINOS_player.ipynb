# Data analysis with NC DINOS player NC 다이노스 선수 데이터 분석

```python
import pandas as pd
```

### 1. 데이터 적재 및 통합

```python
# excel file loading: pd.read_excel('file path')
# optional parameter 'sheet_name=None` > 전체 sheet loading

nc_data = pd.read_excel('data/NC DINOS.xlsx', sheet_name=None)

nc_data
```

```python
# sheet별 data를 각각 개별 변수에 저장
nc_13, nc_14, nc_15 = nc_data.values()
```

```python
# 같은 shape의 data
print(nc_13.shape)
print(nc_14.shape)
print(nc_15.shape)
```

```python
# nc_N data 'Unnamed: 0' column delete

nc_year_data = [nc_13, nc_14, nc_15]

for data in nc_year_data:
    data.drop('Unnamed: 0', axis=1)
    # == del data['Unnamed: 0']
```

```python
# '선수명' columns 이용 > 'nc_13'동명이인이 있는지 여부, 선수명 수 출력
# `.unique()`: 중복값 배제, show 내부 data

print(nc_13['선수명'].count())
print(len(nc_13['선수명'].unique()))
```

```python
# 전체년도 동명이인 여부

for data in nc_year_data:
    if data['선수명'].count() != len(data['선수명'].unique()):
        print(True)
    else:
        print(False)
```

```python
# set data type > 내부 중복 data > item 1개로 간주
# `set()` > 여러 년도 선수 확인 가능
# 13, 14, 15년 선수 수=20 동일
# 14, 15년 연속 주전이었던 선수의 인원 수

len(set(nc_14.선수명.unique()) & set(nc_15['선수명'].unique()))
```
